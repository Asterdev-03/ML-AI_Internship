# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading data

```python
data=pd.read_excel('iris.xls')
data.head()
```

# Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(data['SL'],rwidth=.7,density=True)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('Frequency')
```

```python
data[['SL','PL']].plot.hist()
```

# KDE

```python
data[['SL','PL']].plot.kde()
```

# Scatter plot

```python
plt.figure(figsize=(10,5))
plt.scatter(data['SL'],data['PL'],s=12,marker='o')
plt.title('Plot of SL vs PL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('PL')
```

# Pair plot

```python
sns.pairplot(data)
```

# Heat map

```python
sns.heatmap(data.corr(),annot=True)
plt.yticks(rotation=360)
```

# Box plot

```python
sns.boxplot(x='SL',y='PL',data=data)
```

# Strip plot

```python
 sns.stripplot(data['SL'])
```

# Swarm Plot

```python
sns.swarmplot(x='SL',y='PL',data=data)
```

# Violin plot

```python
sns.violinplot(x='SL',y='PL',data=data)
```

# Count plot

```python
sns.countplot(data['SL'])
plt.grid()
```
