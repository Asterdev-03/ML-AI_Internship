```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# auto-mpg

```python
file1 = pd.read_csv('auto-mpg.csv')
```

```python
file1
```

```python
plt.figure(figsize=(10,5))
plt.hist(file1['mpg'],color='cyan',rwidth=0.9,density=True)
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
file1['mpg'].plot.kde(color = 'blue')
```

```python
file1[['weight','acceleration']].plot.hist(rwidth = 0.8)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(file1['horsepower'],file1['mpg'],s=12,c='lime',marker='+')
plt.title('Plot of Weight vs Acceleration',fontsize=15)
plt.xlabel('Weight')
plt.ylabel('Acceleration')
```

```python
file=file1.drop([32,126, 330, 336, 354, 374])
file['horsepower']=file['horsepower'].astype(float)
```

```python
file
```

```python
file[['weight','acceleration']].plot.hist(rwidth = 0.8)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(file['horsepower'],file['mpg'],s=12,c='lime',marker='+')
plt.title('Plot of Weight vs Acceleration',fontsize=15)
plt.xlabel('Weight')
plt.ylabel('Acceleration')
```

```python
file1.plot(figsize = (10,5))
```
