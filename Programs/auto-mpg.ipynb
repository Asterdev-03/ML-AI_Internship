# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading data

```python
data=pd.read_csv('auto-mpg.csv')
data.head()
```

# Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(data['mpg'],color='darkblue',rwidth=.7,density=True)
plt.title('Distribution of mpg',fontsize=16)
plt.xlabel('MPG')
plt.ylabel('Frequency')
```

### Create data1

```python
data1=data.drop([32,126, 330, 336, 354, 374])
data1['horsepower']=data1['horsepower'].astype(float)
```

# Histogram

```python
plt.hist(data1['mpg'])
plt.hist(data1['horsepower'])
```

# KDE

```python
data1[['mpg','horsepower']].plot.kde()
```


# Scatter plot

```python
plt.figure(figsize=(10,5))
plt.scatter(data1['horsepower'],data1['mpg'],s=12,c='blue',marker='o')
plt.title('Plot of HP vs MPG',fontsize=16)
plt.xlabel('HP')
plt.ylabel('MPG')
```

# Pair plot

```python
sns.pairplot(data)
```

```python
slice_data=data[['mpg','origin','cylinders']]
sns.pairplot(slice_data)
```

# Heat map

```python
sns.heatmap(slice_data.corr(),annot=True)
plt.yticks(rotation=360)
```

# Box plot

```python
sns.boxplot(x='cylinders',y='mpg',data=data)
```

# Strip plot

```python
 sns.stripplot(data['mpg'])
```

# Swarm plot

```python
sns.swarmplot(x='cylinders',y='mpg',data=data)
```

# Violin plot

```python
sns.violinplot(x='cylinders',y='mpg',data=data)
```

# Count plot

```python
sns.countplot(data['origin'])
plt.grid()
```
